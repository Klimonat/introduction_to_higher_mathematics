2. Найдите псевдорешение:

- x + 2y – z = 1
- 3x – 4y = 7
- 8x – 5y + 2z = 12
- 2x – 5z = 7
- 11x +4y – 7z = 15


```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
import scipy.linalg
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-10, 10, 500)
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection='3d')
y = np.linspace(-10, 10, 500)
ax.plot(x, y, x + 2 * y - 1, label='parametric curve 1')
ax.plot(x, (3 * x - 7) / 4, y, label='parametric curve 2')
ax.plot(x, y, (12 - 8 * x + 5 * y ) / 2, label='parametric curve 3')
ax.plot(x, y, (2 * x - 7) / 5, label='parametric curve 4')
ax.plot(x, y, (11 * x + 4 * y - 15) / 7, label='parametric curve 5')
plt.legend()
```

```python
A = np.array([[1, 2, -1], [3, -4, 0], [8, -5, 2], [2, 0 , -5], [11, 4, -7]])
B = np.array([1, 7, 12, 7, 15])
```

```python
np.linalg.lstsq(A, B, rcond=None)
```

```python
np.dot(A, [ 1.13919353, -0.90498444, -0.9009803 ]) - B
```

Ответ: [x, y, z] = [ 1.13919353, -0.90498444, -0.9009803 ]

5. Найдите нормальное псевдорешение недоопределенной системы:
- x + 2y – z = 1  
- 8x – 5y + 2z = 12

Для этого определите функцию Q(x,y,z), равную норме решения, и найдите ее минимум. 

```python
def Q(x, y, z):
    return (x ** 2 + y ** 2 + z ** 2)
```

```python
A_matrix = [[1, 2, -1], [8, -5, 2]]
B_vector = [1, 12]
A_matrix, B_vector
```

```python
np.linalg.lstsq(A_matrix, B_vector, rcond=None)
```

```python
fig = figure(figsize=(10, 10))
ax = Axes3D(fig)
X = np.arange(-50, 50, 1)
Y = np.arange(-50, 50, 1)
X, Y = np.meshgrid(X, Y)
ax.plot_surface(X, Y, Q(X, 10 * X - 14, 21 * X - 29))
plt.show()
```

Ответ: [x, y, z] = [ 1.38191882, -0.18081181,  0.0202952 ]
