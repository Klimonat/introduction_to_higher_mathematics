- Если численным методом не получается найти все ветви решения x(а), то отыщите хотя бы одну.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def func_y(a, x):
    return np.sin(a * x)
```

```python
x = np.linspace(100, 500, 1001)
a_params = np.linspace(0.01, 0.02, 11)

plt.figure(figsize=(16, 8))
for a in a_params:
    y = func_y(a, x)
    plt.plot(x, y, label='a = ' + str(round(a, 4)))
    
plt.legend()
plt.xlabel('Значения переменной x')
plt.ylabel('Значения функции y')
plt.show()
```
