Задание к видеоуроку "Элементарная алгебра"

Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
x = np.linspace(0, 12 * math.pi, 1000)
```

```python
plt.figure(figsize=(10, 6))
k = 1
plt.plot(x, np.cos(k * x), label=('cos(kx), k =' + str(k)), color='c')
k = 1 / 2
plt.plot(x, np.cos(k * x), label=('cos(kx), k =' + str(k)), color='r')
plt.legend(loc=0)
plt.grid()
plt.xlabel("Значения аргумента x")
plt.ylabel("Значения функции y = cos(k * x)")
plt.title('Зависимость функции косинуса от значения аргумента при разных коэффициентах')
plt.show()
```
