Напишите код на Python, реализующий построение графиков:
- окружности,	
- эллипса,
- гиперболы.


```python
def plotting(x, y, x_min, x_max, ymin, y_max):
    plt.figure(figsize=(8, 8))
    axes = plt.gca()
    axes.set_xlim([x_min, x_max])
    axes.set_ylim([ymin, y_max])
    plt.plot(x, y)
    plt.grid()
    plt.xlabel("Значения аргумента x")
    plt.ylabel("Значения функции y")
    plt.show()
```

```python
def make_circle(r, x_0=0, y_0=0, number_points=1000):
    y_pos_right = []
    y_pos_left = []
    y_neg_left = []
    y_neg_right = []
    x_pos = np.linspace(x_0, x_0 + r, number_points)
    x_neg = np.linspace(x_0 - r, x_0, number_points)
    for i in range(0, number_points):
        x_value_pos = x_pos[i]
        y_value_pos_right = math.sqrt(r ** 2 - (x_value_pos - x_0) ** 2) + y_0
        
        x_value_pos = x_pos[::-1][i]
        y_value_neg_right = -math.sqrt(r ** 2 - (x_value_pos - x_0) ** 2) + y_0
        
        x_value_neg = x_neg[::-1][i]
        y_value_neg_left = -math.sqrt(r ** 2 - (x_value_neg - x_0) ** 2) + y_0
        
        x_value_neg = x_neg[i]
        y_value_pos_left = math.sqrt(r ** 2 - (x_value_neg - x_0) ** 2) + y_0    

        y_pos_right.append(y_value_pos_right)
        y_pos_left.append(y_value_pos_left)
        y_neg_left.append(y_value_neg_left)
        y_neg_right.append(y_value_neg_right)

    y = np.append(y_pos_right, [y_neg_right, y_neg_left, y_pos_left])
    x = np.append(x_pos, [x_pos[::-1], x_neg[::-1], x_neg])
    return x, y
```

```python
x_0 = 1
y_0 = 1
r = 5
number_points = 1000
x_min = x_0 - r - 1
x_max = x_0 + r + 1
y_min = y_0 - r - 1
y_max = y_0 + r + 1
```

```python
x, y = make_circle(r, x_0, y_0, number_points)
plotting(x, y, x_min, x_max, y_min, y_max)
```

```python
def make_ellipse(a, b, x_0=0, y_0=0, number_points=1000):
    y_pos_right = []
    y_pos_left = []
    y_neg_left = []
    y_neg_right = []
    x_pos = np.linspace(x_0, x_0 + a, number_points)
    x_neg = np.linspace(x_0 - a, x_0, number_points)
    for i in range(0, number_points):
        x_value_pos = x_pos[i]
        y_value_pos_right = b * math.sqrt(a ** 2 - (x_value_pos - x_0) ** 2) / a + y_0
        
        x_value_pos = x_pos[::-1][i]
        y_value_neg_right = - b * math.sqrt(a ** 2 - (x_value_pos - x_0) ** 2) / a + y_0
        
        x_value_neg = x_neg[::-1][i]
        y_value_neg_left = - b * math.sqrt(a ** 2 - (x_value_neg - x_0) ** 2) / a + y_0
        
        x_value_neg = x_neg[i]
        y_value_pos_left = b * math.sqrt(a ** 2 - (x_value_neg - x_0) ** 2) / a + y_0

        y_pos_right.append(y_value_pos_right)
        y_pos_left.append(y_value_pos_left)
        y_neg_left.append(y_value_neg_left)
        y_neg_right.append(y_value_neg_right)

    y = np.append(y_pos_right, [y_neg_right, y_neg_left, y_pos_left])
    x = np.append(x_pos, [x_pos[::-1], x_neg[::-1], x_neg])
    return x, y
```

```python
a = 5
b = 1
x_min = x_0 - a - 1
x_max = x_0 + a + 1
y_min = y_0 - b - 1
y_max = y_0 + b + 1
```

```python
x, y = make_ellipse(a, b, x_0, y_0, number_points)
plotting(x, y, x_min, x_max, y_min, y_max)
```

```python
def make_hyperbole(a, b, x_0=0, y_0=0, number_points=1000):
    y_pos_right = []
    y_pos_left = []
    y_neg_left = []
    y_neg_right = []
    x_pos = np.linspace(x_0 + a, x_0 + 10, number_points)
    x_neg = np.linspace(x_0 - 10, x_0 - a, number_points)
    for i in range(0, number_points):
        x_value_pos = x_pos[i]
        y_value_pos_right = b * math.sqrt((x_value_pos - x_0) ** 2 - a ** 2) / a + y_0
        
        x_value_pos = x_pos[::-1][i]
        y_value_neg_right = - b * math.sqrt(abs((x_value_pos - x_0) ** 2 - a ** 2)) / a + y_0
        
        x_value_neg = x_neg[::-1][i]
        y_value_neg_left = - b * math.sqrt((x_value_neg - x_0) ** 2 - a ** 2) / a + y_0
        
        x_value_neg = x_neg[i]
        y_value_pos_left = b * math.sqrt((x_value_neg - x_0) ** 2 - a ** 2) / a + y_0

        y_pos_right.append(y_value_pos_right)
        y_pos_left.append(y_value_pos_left)
        y_neg_left.append(y_value_neg_left)
        y_neg_right.append(y_value_neg_right)

    y_left = np.append(y_neg_left[::-1], y_pos_left[::-1])
    y_right = np.append(y_pos_right[::-1], y_neg_right[::-1])
    x_left = np.append(x_pos[::-1], x_pos)
    x_right = np.append(x_neg, x_neg[::-1])
    return y_left, y_right, x_left, x_right
```

```python
a = 2
b = 2
x_0 = 0
y_0 = 0
number_points = 1000
x_min = x_0 - a - 10
x_max = x_0 + a + 10
y_min = y_0 - b - 10
y_max = y_0 + b + 10
```

```python
y_left, y_right, x_left, x_right = make_hyperbole(a, b, x_0, y_0, number_points)
```

```python
plt.figure(figsize=(8, 8))
axes = plt.gca()
axes.set_xlim([x_min, x_max])
axes.set_ylim([y_min, y_max])
plt.plot(x_left, y_left)
plt.plot(x_right, y_right)
plt.grid()
plt.xlabel("Значения аргумента x")
plt.ylabel("Значения функции y")
plt.show()
```

Решите систему уравнений и неравенств:

    2) y = x2 – 1
    exp(x) + x∙(1 – y)  - 1 > 0


```python
from scipy.optimize import fsolve


import warnings

warnings.filterwarnings('ignore')
```

```python
x = np.linspace(-3, 3, 201)
plt.plot(x, ((x - np.exp(x) + 1) / x), label='y = (x - np.exp(x) + 1) / x')
plt.plot(x, (x ** 2 - 1), label = 'y = x ** 2 - 1')
plt.legend()
plt.grid()
plt.show()
```

```python
def equations(p):
    x, y = p
    return (y - x ** 2 + 1, np.exp(x) + x * (1 - y) - 1)
```

```python
x_1, y_1 = fsolve(equations, (1, 0.5))
x_1, y_1
```

```python
x_2, y_2 = fsolve(equations, (-1, 0))
x_2, y_2
```

```python
x_3, y_3 = fsolve(equations, (10, 10))
x_3, y_3
```
